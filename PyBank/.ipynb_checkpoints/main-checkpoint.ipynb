```python
# this is a function that takes a list and calulates the average differences between the values of that list
def average_change_list(list_of_values):
    # create empty list to store differences
    difference_list = []
    # empty int to store value in list
    save_value = 0
    #empty int to store sum of the values in the list
    sum_values = 0
    #for loop to add the difference between the values to a list
    for value in list_of_values:

        if save_value == 0:
            save_value = value
        
        else: 
            
            difference = value - save_value
            
            difference_list.append(difference) 
            
            save_value = value
    #for loop to caluclate the sum of the values in the list
    for value in difference_list:
        sum_values += value 
    #finding length of list to determine how much to divide the sum by
    amount = len(difference_list)
    #calculating sum of the list by dividing the sum of differences by amount of differences
    average = round(sum_values/amount, 2) 
    return average  

```

```python
#this is a function that will take a list of values and the months that corespond to them and return a list of the minimum and 
#maximum value as well as the months that correspond with them
def lowest_and_highest_values(list_of_values, correlating_month):
    # create empty list to store differences
    difference_list = []
    # empty int to store value in list
    save_value = 0
    #empty int to store sum of the values in the list
    sum_values = 0
    #for loop to add the difference between the values to a list
    for value in list_of_values:

        if save_value == 0:
            save_value = value
             
        else: 
            
            difference = value - save_value
            
            difference_list.append(difference) 
            
            save_value = value
    #finding the minimum value in the list of differences
    minimum_value = min(difference_list) 
    #finding the month in which the minimum value occurred
    minimum_month = correlating_month[list_of_values.index(min(list_of_values))]
    #finding the maximum value in the list of differences
    maximim_value = max(difference_list)
    #finding the month in which the maximum value occurred
    maximum_month = correlating_month[list_of_values.index(max(list_of_values))] 

    return [minimum_month, minimum_value, maximum_month, maximim_value] 
  
```

```python
# import the pathlib and csv library
from pathlib import Path
import csv 

#set the file path 
csvpath = Path('../budget_data.csv') 

#set empty lists
all_months = []
all_values = []

#set empty variables
line_num = 0
total_months = 0
total_amount = 0 

#open csv file path
with open(csvpath, 'r') as csvfile:


    #read csv file
    csvreader = csv.reader(csvfile)

    #add values and months to a list
    for row in csvreader:
        
        month = str(row[0])

        value = str(row[1])
        
        all_months.append(month)

        all_values.append(value)
   
#remove headers from the lists
all_months.pop(0)
all_values.pop(0)
#transform all variables to integers
all_values = list(map(int,all_values)) 


#calculate total amount of months
total_months = len(all_months)
#calculate total sum of all values 
for items in all_values:
    total_amount += items 
#call function fom above to return average change
average = average_change_list(all_values)
#call function from above to return list of maximum and minimum values and months
max_and_mim = lowest_and_highest_values(all_values, all_months)
```

```python
#All Outputs

print("Financial Analysis:")
print("----------------------------")
print(f"Total Months: {total_months}")
print(f"Total: ${total_amount}")
print(f"Average Change: ${average}")
print(f"Greatest Increase in Profits: {max_and_mim[2]}  (${max_and_mim[3]})")
print(f"Greatest Decrease in Profits: {max_and_mim[0]} (${max_and_mim[1]})") 


output_path = "output.txt"


with open(output_path, 'w') as file:
    file.write("This is output file. /n")
    file.write("Financial Analysis:")
    file.write(f"Total Months: {total_months}")
    file.write(f"Total: ${total_amount}")
    file.write(f"Average Change: ${average}")
    file.write(f"Greatest Increase in Profits: {max_and_mim[2]}  (${max_and_mim[3]})")
    file.write(f"Greatest Decrease in Profits: {max_and_mim[0]} (${max_and_mim[1]})")
    file.write("Financial Analysis:")
```
